```python
# Core dependencies
import numpy as np
import scipy
import sklearn
import cPickle as pickle

import librosa
import gordon
```

```python
# My own online learning modules
# Learn OOC comes from https://github.com/bmcfee/learn_ooc
# The rest come from https://github.com/bmcfee/ml_scraps

# Learn OOC is the glue that allows sequential learning (sklearn) from generator functions
# Whitening, Hartigan, and VQ are all sklearn modules for whitening transformations, 
# online clustering, and vector quantization
import learn_ooc
import Whitening, HartiganOnline, VectorQuantizer
```

```python
# Get all the tracks corresponding to a named collection in gordon
def get_collection_tracks(name):
    C = gordon.Collection.query.filter(gordon.Collection.name==name).limit(1)
    C = C.all()
    if len(C) == 0:
        return []
    return [t.id for t in C[0].tracks]
```

```python
# Get the track object by id
def get_track(t_id):
    t = gordon.Track.query.filter(gordon.Track.id==t_id).limit(1)
    t = t.all()
    return t[0]
```

```python
# get the analysis data for a track by id
def get_analysis(track_id):
    
    track = get_track(track_id)
    
    analysis_file = track.annotation_dict['librosa:low-level']
    
    with open(analysis_file, 'r') as f:
        data = pickle.load(f)
        
    return data
```

```python
# Get local contrast spectrogram features from a track id
def extract_features(track_id):
    
    # Build the local difference mapping
    def make_contrast_features(X, widths=[3, 7, 15]):
    
        d = X.shape[0]
        Xhat = X
    
        for w in widths:
            Xhat = np.vstack((Xhat, scipy.signal.lfilter(np.ones(w) / w, 1, X, axis=-1)))
        
        return Xhat
    
     # Get the mfcc's
    S = get_analysis(track_id)['mel_spectrogram']
    S = librosa.logamplitude(S, ref_power=S.max())
    
    return make_contrast_features(S)
```

```python
# The datastream object's feature generate
# Takes a track id, and generates samples of its feature content
def mapper(track_id, n=50, whiten=None):
    
    # Get the features
    features = extract_features(track_id)
    
    if whiten is not None:
        features = whiten.transform(features.T).T
    
    for i in range(n):
        t = np.random.randint(0, features.shape[1])
        yield features[:, t]
```

```python
# Get tracks from the ffuhrmann collection.  
# We'll just use the first 1000 for now
ffuhrmann_tracks = get_collection_tracks('ffuhrmann')[:1000]
```

```python
# Create a data stream to learn a whitening transformer
data_stream = learn_ooc.mux_bounded(mapper, 
                                    [t for t in ffuhrmann_tracks], 
                                    working_size=256, 
                                    max_iter=1, 
                                    n=20)
```

```python
# Build the whitening transform
est_white = Whitening.Whitening()
learn_ooc.fit(est_white, data_stream, batch_size=1024)
```

```python
# Create a new data stream that uses the whitener prior to running k-means
# This could also be done with a sklearn.pipeline, probably?
data_stream = learn_ooc.mux_bounded(mapper, 
                                    [t for t in ffuhrmann_tracks], 
                                    working_size=128, 
                                    max_iter=3, 
                                    n=64, 
                                    whiten=est_white)
```

```python
# Build the codebook estimator. 
# I'll use my online hartigan for the codebook learning

estimator = VectorQuantizer.VectorQuantizer(clusterer=HartiganOnline.HartiganOnline(n_clusters=2048))
learn_ooc.fit(estimator, data_stream, batch_size=1024)
```

```python
# What's the codebook look like?
# We'll sort the columns by peak feature
figure(figsize=(12, 8))
librosa.display.specshow(librosa.util.axis_sort(estimator.components_.T))
colorbar()
xlabel('Codewords')
ylabel('Features')
tight_layout()
```

```python
# Let's plot some codewords as patches.
# Each sub-block of a codeword corresponds to an averaging-window length.
figure(figsize=(12, 6))
k = 16
cmap = librosa.display.cmap(estimator.components_)
for i in range(1, k+1):
    subplot(1, k, i)
    j = np.random.randint(0, estimator.components_.shape[0])
    cw = np.reshape(estimator.components_[j], (4, -1))
    librosa.display.specshow(cw[::-1].T, cmap=cmap)
    xticks([0,1,2,3], [-15, -7, -3, 0], rotation=90)
    title(j)
tight_layout()
```
